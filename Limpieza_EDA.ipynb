## 1. Eliminar las columnas cuyos datos que tengan el 75% de nulls en adelante, ya que no son representativos

```python
#porcentaje de null  en regla de 3 simple
def getNullPercent(columnName):
  #columnName='semana'
  totalNulls=df[columnName].isna().sum()
  totalRegistros=len(df[columnName])
  nullPercent=100*totalNulls/totalRegistros
  return nullPercent
```

```python
#probamos con dos registros
print(round(getNullPercent('id'),2))
print(round(getNullPercent('entrada'),2))
```

```python
# columnas a remover
def getColumsToDelete():

  columnsDelete=[]
  for column in df.columns:
    nameColumn=column
    percentNull=round(getNullPercent(column),2)
    if(percentNull>=75):
      #print(f"{nameColumn}:   {percentNull}% -->será removido")
      columnsDelete.append(nameColumn)
    else:
      #print(f"{nameColumn}:   {percentNull}%")
      pass
  return columnsDelete
```

```python
print(getColumsToDelete())
```

```python
df=df.drop(columns=getColumsToDelete())
```

```python
# verificamos las columnas con registros numéricos
def getColumnsNumeric():
  columnsNumeric=[]
  import numpy as np
  for columna in df.columns:

      #si la columna es numérica
      if np.issubdtype(df[columna].dtype, np.number):

          columnsNumeric.append(columna)
          # Calcular la media de la columna
          media = df[columna].mean()
          print(f"La columna '{columna}' es numérica. La media es {round(media,2)}")

          #si la columna tiene valores nulos
          if df[columna].isnull().values.any():
            # Reemplazar los valores nulos con la media
            print(f"Sus valores nulos de esta columna {columna} serán reemplazados por su media")
            df[columna].fillna(media, inplace=True)
  return columnsNumeric

```
